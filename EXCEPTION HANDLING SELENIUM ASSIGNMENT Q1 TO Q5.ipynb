```python
driver=webdriver.Chrome(r"chromedriver.exe")
driver.get("https://www.flipkart.com/")                                                      ## ANSWER Q-4
```

```python
urls1 = []
for i in driver.find_elements(By.XPATH,'//input[@class="_3704LK"]'):
    urls1.append(i.get_attribute("href"))
```

```python
print(len(urls1))
```

```python
smartphone_name = [] 
price = [] 
ratings = [] 
description = [] 

for i in urls1: 
    try:
        name = driver.find_elements(By.XPATH,'//div[@class="_4rR01T"]')
        smartphone_name.append(name.text)
    except:
        smartphone_name.append('-')         
    try:
        price1 =  driver.find_elements(By.XPATH,'//div[@class="_30jeq3 _1_WHN1"]')
        price.append(price1.text)
    except:
        price.append('-')   
    try:
        rate = driver.find_elements(By.XPATH,'//span[@class="_2_R_DZ"]')
        ratings.append(rate.text)
    except:
        ratings.append('-')
    try:
        desc= driver.find_elements(By.XPATH,'//div[@class="fMghEO"]')
        description.append(desc.text)
    except:
        description.append('-') 
```

```python
len(smartphone_name)
len(price)
len(ratings)
len(description)
```

```python
mobiles = pd.DataFrame({})

mobiles['Name of the smartphone']= smartphone_name
mobiles['Price']= price
mobiles['Ratings and reviews']= ratings
mobiles['url of the page']= urls1
```
